# **Matriz de confusión**
Ayuda a evaluar el rendimiento del modelo de clasificación en el aprendizaje automático

**Estructura**
*  **Positivo** (P): La observación es positiva (por ejemplo, es un perro).
*  **Negativo** (N): La observación no es positiva (por ejemplo, no es un perro).
*  **Verdadero positivo** (TP) - Tu modelo predijo la clase positiva. Por ejemplo, identificar un correo electrónico como spam.
* **Negativo verdadero** (NT) - Tu modelo predijo correctamente la clase negativa. Por ejemplo, identificar un correo electrónico normal como no spam.
* **Falso positivo** (FP) -  Error de tipo 1, Tu modelo predijo incorrectamente la clase positiva. Por ejemplo, identificar un correo normal como spam.
* **Falso negativo** (FN) - Error de tipo 2, Tu modelo predijo incorrectamente la clase negativa. Por ejemplo, identificar un correo basura como un correo normal.


1.   **Exactitud**

  Mide cuántas predicciones fueron correctas en total.

  Exactitud = 𝑇𝑃 + 𝑇𝑁  / 𝑇𝑃 + 𝑇𝑁 + 𝐹𝑃 + 𝐹𝑁

2.   **Precisión**

  De todos los que el modelo dijo que eran positivos, ¿cuántos realmente lo eran?

  Precisión= TP / TP+FP

3.   **Sensibilidad**

  De todos los casos realmente positivos, ¿cuántos detectó el modelo?

  Sensibilidad= TP / TP+FN

4.   **Especificidad**

  De todos los casos realmente negativos, ¿cuántos clasificó correctamente el modelo?
  
  Especificidad= TN / TN+FP

5.   **Puntuación F1**

  Es el promedio armónico entre precisión y sensibilidad.

  F1 = 2TP / 2TP + FP + FN
